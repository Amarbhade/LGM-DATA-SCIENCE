## TASK 2= IMAGE TO PENCIL SKETCH WITH PYTHON

### IMPORTING REQUIRED LIBRARY

```python
# the open cv2 is the library for computr vision machine learning and image processing
import cv2
```

```python
from matplotlib import pyplot as plt
```

#### for reading image

```python
picture=cv2.imread("kitten.png")
```

```python
# for showing image

def show_picture(picture):
    plt.figure(dpi=100)
    read_picture=cv2.cvtColor(picture , cv2.COLOR_BGR2RGB)
    plt.imshow(read_picture)
    plt.axis("off")
    plt.show()
```

```python
show_picture(picture)
```

### color image converted into gray

```python
# for converting into colorful to gray
gray_picture= cv2.cvtColor(picture, cv2.COLOR_BGR2GRAY)
```

```python
#import library
import matplotlib as mat
# for picture output
plt.figure(dpi=100)
plt.imshow(gray_picture,cmap=mat.cm.binary_r);
plt.axis("off");
```

```python
# now we convert image into the dark gray
dark_gray_picture=cv2.bitwise_not(gray_picture)
# for picture output
plt.figure(dpi=100)
plt.imshow(dark_gray_picture,cmap=mat.cm.binary_r);
plt.axis("off");
```

```python
# for smoothing the picture 
smooth_picture=cv2.GaussianBlur(dark_gray_picture ,(19,21),0)
# for showing
plt.figure(dpi=100)
plt.imshow(smooth_picture ,cmap=mat.cm.binary_r);
plt.axis("off")
```

### now we converted picture into the pencil sketch

```python
# for pencil sketch

picture_sketch= 253- smooth_picture
pencil_sketch= cv2.divide(gray_picture ,picture_sketch ,scale= 246.0)
# fow show
plt.figure(dpi=100)
plt.imshow(pencil_sketch ,cmap=mat.cm.binary_r)
("here we convert picture into pencil sketch")
plt.axis("off")
```
